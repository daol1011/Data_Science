#### Exercise 1.1a)
We compute R(A | G) and compare the acceptance rate between men and women:

```python
pd.crosstab(d['Geschlecht'],columns=d['Angenommen'], margins=True)
```

```python
pd.crosstab(d['Geschlecht'],columns=d['Angenommen'], margins=True, normalize='all')
```

```python
pd.crosstab(d['Geschlecht'],columns=d['Angenommen'], normalize='index', margins=True)
```

#### Answer 1.1a:
Around 48% of men, but only about 33% auf women got accepted. The variables "gender" and "acceptance" are therefore statistically *dependent*.

##### OR:
R( A=1 | G='M') = 48%  !=  33% = R(A=1 | G='W') . A and G are therefore statistically *dependent*.

#### Exercise 1.1b)
We repeat the acceptance rate comparison from (a), but this time separately for the different subjects. To do this, we calculate R(A | G,F):

```python
pd.crosstab([d['Fach'], d['Geschlecht']],columns=d['Angenommen'],margins=True)
```

```python
pd.crosstab([d['Fach'], d['Geschlecht']],columns=d['Angenommen'], normalize='index')
```

#### Answer 1.1b
For each of the subjects, the acceptance rate is the same for women and men. The different subject preferences between women and men therefore fully explain the dependency between gender and acceptance.   

Gender is useless for predicting the acceptance rate, given the subject

##### OR:
R(A=1 | Fach='Informatik', G='M') = 0.6 = R(A=1 | Fach='Informatik', G='W'),
R(A=1 | Fach='IntMgmt', G='M') = 0.2 = R(A=1 | Fach='IntMgmt', G='W'),
R(A=1 | Fach='WI', G='M') = 0.5 = R(A=1 | Fach='WI', G='W'),
Under the condition that the subject is known (or also "given the subject"), A and G are therefore statistically independent, i.e. F fully explains the relationship between G and A.

##### Plausibility check:
Of course this makes us curios whether the subject preferences between women and men are truly so different. We therefore calculate the subject distribution for women and men separately, i.e. R(F | G):

```python
pd.crosstab(d['Geschlecht'],columns=d['Fach'], normalize='index')
```

```python
pd.crosstab(d['Fach'],columns=d['Angenommen'], normalize='index')
```

Indeed, a much higher proportion of women (approx. 63%) apply for IntMgmt, where the acceptance rate is very low, than is the case for men (approx. 24%).
