```python
crosstab_result = pd.crosstab(index=d['Geschlecht'],columns=d['Angenommen'], normalize='index')

print(crosstab_result)

conditional_prob_acc = crosstab_result[1]
conditional_prob_den = crosstab_result[0]
print("The probability of being accepted if we are male is ", conditional_prob_acc['M'])
print("The probability of being accepted if we are female is ", conditional_prob_acc['W'])
print("The probability of being denied if we are male is ", conditional_prob_den['M'])
print("The probability of being denied if we are female is ", conditional_prob_den['W'])
print("Thus we can see there is a stochastic dependance of gender and acceptance rate, being male slightly increases our change of being accepted")
```
